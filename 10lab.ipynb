```python
import pandas as pd
data = pd.read_csv('gun-violence-data.csv')
data = data.drop(columns=['incident_id', 'date', 'state', 'city_or_county', 'notes', 'participant_status',
                          'participant_type', 'participant_gender', 'gun_type', 'participant_age_group',
                          'participant_name', 'address', 'incident_url', 'source_url', 'gun_stolen', 
                          'incident_url_fields_missing', 'sources', 'participant_age'])
data = data.dropna()
data.head()
```

```python
data['location_description'] = data['location_description'].str.lower()
```

```python
import re

def f(x):
    if re.search('pharm|shop|store|market|grocery|walmart|7|jewel|mall|dollar|product|circle|plaza', x) != None:
        return 0
    elif re.search('bank|company| inc|center|office| co|group', x) != None:
        return 1
    elif re.search('gas|bp|shell|exxon|auto|car|station|fuel|chevron|citgo', x) != None:
        return 2
    elif re.search('food|cafe|bar|donut|burger|pub|buffet|grill|mcdonald|kf|\'s|pizza|restaurant|sandwic|taco', x) != None:
        return 3
    elif re.search('hotel| inn|motel|camp|spa', x) != None:
        return 4 
    elif re.search('subway|bridge|park|stop|forest|beach|way|road', x) != None:
        return 5
    elif re.search('apartments|house|vill|reservation|yard', x) != None:
        return 6
    elif re.search('university|college|library|school|club|church|hospital|elementary|theater', x) != None:
        return 7
    elif re.search('east|west|north|south|district|neighborhood|area|central|township|hill|land', x) != None:
        return 8
    return 9

data['location_description'] = data['location_description'].apply(f)
data.head()
```

```python
def t(x):
    if re.search('Armed Robbery', x) != None:
        return 0
    elif re.search('Family', x) != None:
        return 1
    elif re.search('Aquaintance', x) != None:
        return 2
    elif re.search('Significant others - current or former', x) != None:
        return 3
    elif re.search('Friends', x) != None:
        return 4 
    elif re.search('Neighbor', x) != None:
        return 5
    elif re.search('Co-worker', x) != None:
        return 6
    elif re.search('Gang vs Gang', x) != None:
        return 7
    elif re.search('Home Invasion - Perp Does Not Know Victim', x) != None:
        return 8
    elif re.search('Mass shooting - Random victims', x) != None:
        return 9
    elif re.search('Home Invasion - Perp Knows Victim', x) != None:
        return 10
    return 11

data['participant_relationship'] = data['participant_relationship'].apply(t)
data.head()
```

```python
def a(x):
    val = 0
    if re.search('Accidental Shooting', x) != None:
        val = val + 10
    if re.search('Defensive Use', x) != None:
        val = val + 20
    if re.search('Armed robbery', x) != None:
        val = val + 30
    
    return val

data['is_child'] = data['incident_characteristics'].apply(a,axis=1)
data.head()
```

# В кубе 100 товаров, 10 городов и 30 дней

```python
import numpy as np

print("В кубе 100 товаров, 10 городов и 30 дней")

k_items = np.random.exponential(100,size=100)
cube = np.array([[ [int(k_items[i]*np.random.random())
                   for k in range(30) ] for j in range(10)]for i in range(100)])

items = np.sort(np.random.randint(0,100,4))
cities = np.sort([2, 6, 8, 5])

print("Необходимо посчитать значения для товаров {0} "
      "в период 15-26 дни по городам {1}".format(items, cities))
```

## Суммарные значения

```python
for item in items:
    print("\nТовар {0}".format(item))
    for city in cities:
        print("В городе {0} продано {1}".format(city,cube[item, city, 14:25].sum()))
    print("Товара {0} продано во всех городах {1}".format(item, cube[item, cities.tolist(),
                                                          14:25].sum()))
print("Всех товаров продано " + cube[items.tolist(), cities.tolist(), 14:25].sum().__str__())
```

## Средние значения

```python
for item in items:
    print("\nТовар {0}".format(item))
    for city in cities:
        print("В городе {0} продано в среднем {1}".format(city,cube[item, city, 14:25].mean()))
    print("Товара {0} продано в среднем во всех городах {1}".format(item,
                                                                    cube[item, cities.tolist(), 14:25].mean()))
print("Всего товаров в среднем " + cube[items.tolist(), cities.tolist(), 14:25].mean().__str__())
```

## Минимальные значения

```python
for item in items:
    print("\nТовар {0}".format(item))
    for city in cities:
        print("В городе {0} минимум продано {1}".format(city,np.min(cube[item, city, 14:25])))
    print("Товара {0} минимум продано во всех городах {1}".format(item,
                                                                  cube[item, cities.tolist(), 14:25].min()))
print("Общий минимум " + cube[items.tolist(), cities.tolist(), 14:25].min().__str__())
```

## Максимальные значения

```python
for item in items:
    print("\nТовар {0}".format(item))
    for city in cities:
        print("В городе {0} макисмум продано {1}".format(city,np.max(cube[item, city, 14:25])))
    print("Товара {0} максимум продано во всех городах {1}".format(item,
                                                                   cube[item, cities.tolist(), 14:25].max()))
print("Общий максимум " + cube[items.tolist(), cities.tolist(), 14:25].max().__str__())
```
